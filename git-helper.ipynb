# Git helper notebook
Use this notebook to inspect git status and optionally commit/push changes.

```python
import subprocess
from pathlib import Path
root = Path('d:/Gabin/Webify')
print('cwd=', root)
print('git version:')
print(subprocess.check_output(['git', '--version'], cwd=root, text=True))
print('status:')
print(subprocess.check_output(['git', 'status', '--short', '--branch'], cwd=root, text=True))
```
